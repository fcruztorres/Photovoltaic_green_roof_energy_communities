```python
overview_data = pd.read_csv("Costs and benefits of photovoltaic-green roofs - selected benefits.csv")

costs = overview_data[(overview_data["Item type"]=="cost") & 
                      (overview_data["Base case roof"]=="black roof") & 
                      (overview_data["Item"]!="Installation of green roof (gross)") & 
                      (overview_data["Item"]!="Maintenance of green roof (gross)") & 
                      (overview_data["Item"]!="Maintenance of green roof") &
                     (overview_data["Item"]!="Maintenance of green roof (net)")]
costs.boxplot(column="Adjusted Monetary value [EU27 2019 €/m2]", by=["Item"], figsize=(9,6), rot=90, fontsize= 15, showfliers=True)
plt.ylabel('Adjusted Monetary value [EU27 2019 €/m2]', fontsize=14)
plt.title('', fontsize=15)
plt.savefig("Figures/Overview of one-time costs",dpi=300, bbox_inches = "tight")
```

```python
maintenance = overview_data[(overview_data["Item"]=="Maintenance of green roof (net)")]
maintenance.boxplot(column="Adjusted Monetary value [EU27 2019 €/m2]", by=["Item"], figsize=(4,6), rot=90, fontsize= 15, showfliers=True)
plt.ylabel('Adjusted Monetary value [EU27 2019 €/m2/year]', fontsize=14)
plt.title('', fontsize=15)
plt.savefig("Figures/Maintenance",dpi=300, bbox_inches = "tight")
```

```python
costs.groupby(["Item"])["Adjusted Monetary value [EU27 2019 €/m2]"].describe()
```

```python
# Distribution summary statistics of maintenance costs: 
# unfortunately cost items without irrigation are too few to draw general conclusions.
costs[costs["Item"]=="Maintenance of green roof (net)"].groupby("Irrigation")["Adjusted Monetary value [EU27 2019 €/m2]"].describe()
```

```python
benefits = overview_data[(overview_data["Item type"]=="benefit") &
                        (overview_data["Base case roof"]=="black roof") ]
benefits.boxplot(column="Adjusted Monetary value [EU27 2019 €/m2]", by=["Item"], figsize=(20,6), rot=45, showfliers=True)
plt.ylabel('Adjusted Monetary value [EU27 2019 €/m2]', fontsize=14)
plt.title('', fontsize=15)
plt.savefig("Figures/Overview of the benefits",dpi=300, bbox_inches = "tight")
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(50,10)) # create 1x2 array of subplots

benefits.boxplot(column="Adjusted Monetary value [EU27 2019 €/m2]", by=["Item"], figsize=(20,6), rot=45, fontsize= 15, showfliers=False, ax=axes[0])
plt.title('Adjusted Monetary value [EU27 2019 €/m2]', fontsize=15)

overview_data[overview_data["Item"]=="Aesthetics increase"].boxplot(column="Adjusted Monetary value [EU27 2019 €/m2]", by=["Item"], figsize=(5,6), rot=0, fontsize= 15, showfliers=False, ax=axes[1])
plt.title('Adjusted Monetary value [EU27 2019 €/m2]', fontsize=15)
plt.tight_layout
plt.show()
```
