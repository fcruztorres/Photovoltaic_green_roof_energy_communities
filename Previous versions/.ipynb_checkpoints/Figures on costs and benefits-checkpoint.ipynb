```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
raw_data = pd.read_csv("Costs and benefits of photovoltaic-green roofs.csv")
raw_data
```

```python
raw_data["Item"].unique()
```

```python
air_quality_issues = raw_data[(raw_data["Item"]=="Air quality enhancement") |
                              (raw_data["Item"]=="CO2 emission reduction") |
                              (raw_data["Item"]=="CO2 uptake") ]
air_quality_issues.boxplot(column="Adjusted Monetary value [EU27 2019 €/m2]", by=["Item"], figsize=(10,6), showfliers=False)
plt.savefig("Figures/Air quality overlaps")
```

```python
air_quality_issues.groupby(["Item"])["Adjusted Monetary value [EU27 2019 €/m2]"].describe()
```

```python
sound_aesth_prop_issues = raw_data[(raw_data["Item"]=="Sound insulation") |
                              (raw_data["Item"]=="Aesthetics increase") ]

sound_aesth_prop_issues.boxplot(column="Adjusted Monetary value [EU27 2019 €/m2]", by=["Item"], figsize=(10,6), showfliers=True)
plt.savefig("Figures/Property value increase overlaps")
```

```python
sound_aesth_prop_issues.groupby(["Item"])["Adjusted Monetary value [EU27 2019 €/m2]"].describe()
```

```python
PV_perf_increase = raw_data[(raw_data["Item"]=="Photovoltaic output")]

PV_perf_increase.boxplot(column="Adjusted Monetary value [EU27 2019 €/m2]", by=["Item"], figsize=(4,6), showfliers=True)

plt.savefig("Figures/PV performance")
```

```python
PV_perf_increase = raw_data[(raw_data["Item"]=="Photovoltaic performance increase") |
                           (raw_data["Item"]=="Photovoltaic output")]

PV_perf_increase.boxplot(column="Adjusted Monetary value [EU27 2019 €/m2]", by=["Item"], figsize=(10,6), showfliers=True)
plt.savefig("Figures/PV")
```

```python
overview_data = pd.read_csv("Costs and benefits of photovoltaic-green roofs - selected benefits.csv")

costs = overview_data[(overview_data["Item type"]=="cost") & 
                      (overview_data["Base case roof"]=="black roof") & 
                      (overview_data["Item"]!="Installation of green roof (gross)") & 
                      (overview_data["Item"]!="Maintenance of green roof (gross)") & 
                      (overview_data["Item"]!="Maintenance of green roof") ]
costs.boxplot(column="Adjusted Monetary value [EU27 2019 €/m2]", by=["Item"], figsize=(20,6), rot=45, fontsize= 15, showfliers=True)
plt.title('Adjusted Monetary value [EU27 2019 €/m2]', fontsize=15)
plt.savefig("Figures/Overview of the costs")
```

```python
costs.groupby(["Item"])["Adjusted Monetary value [EU27 2019 €/m2]"].describe()
```

```python
# Distribution summary statistics of maintenance costs: 
# unfortunately cost items without irrigation are too few to draw general conclusions.
costs[costs["Item"]=="Maintenance of green roof (net)"].groupby("Irrigation")["Adjusted Monetary value [EU27 2019 €/m2]"].describe()
```

```python
benefits = overview_data[(overview_data["Item type"]=="benefit") &
                        (overview_data["Base case roof"]=="black roof") ]
benefits.boxplot(column="Adjusted Monetary value [EU27 2019 €/m2]", by=["Item"], figsize=(20,6), rot=45, showfliers=True)
plt.savefig("Figures/Overview of the benefits")
```

```python
benefits = overview_data[(overview_data["Item type"]=="benefit") & 
                         (overview_data["Base case roof"]=="black roof") & 
                         (overview_data["Item"]!="Aesthetics increase") &
                         (overview_data["Item"]!="Air quality enhancement") &
                        (overview_data["Item"]!="CO2 emission reduction") &
                        (overview_data["Item"]!="CO2 uptake") ]
fig = benefits.boxplot(column="Adjusted Monetary value [EU27 2019 €/m2]", by=["Item"], figsize=(20,6), rot=90, fontsize= 15, showfliers=False)
plt.title('Adjusted Monetary value [EU27 2019 €/m2]', fontsize=15)
plt.savefig("Figures/Zoom on smaller benefits")
```
