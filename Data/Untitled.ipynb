```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def inter_row_distance(solar_azimuth, sun_elevation):
    return np.cos(solar_azimuth/360*2*np.pi)/np.tan(sun_elevation/360*2*np.pi)
```
